```python
# สร้าง dictionary สำหรับการแมป
age_mapping = {
    'ต่ำกว่า 18 ปี': 0,
    '18-24 ปี': 1,
    '25-34 ปี': 2,
    '35-44 ปี': 3,
    '45-54 ปี': 4,
    '55-64 ปี': 5
}

# แปลงข้อมูลในคอลัมน์ 'อายุ : ' ด้วยการใช้ dictionary
df['อายุ_encoded'] = df['อายุ : '].map(age_mapping)
df['อายุ_encoded'] = df['อายุ_encoded'].astype(float)
# ตรวจสอบผลลัพธ์
print(df[['อายุ : ', 'อายุ_encoded']].head())


# สร้าง dictionary สำหรับการแมป
impact_mapping = {
    'ไม่มีผลกระทบ': 0,
    'มีผลดี (เช่น ได้ความรู้, คลายเครียด)': 1,
    'มีผลเสีย (เช่น ทำให้นอนดึก, ขาดสมาธิ)': 2,
    'ทั้งผลดีและผลเสีย': 3
}

# แปลงข้อมูลในคอลัมน์ 'คุณคิดว่าการเสพสื่อของคุณมีผลกระทบต่อการใช้ชีวิตประจำวันหรือไม่?' ด้วยการใช้ dictionary
df['ผลกระทบ_encoded'] = df['คุณคิดว่าการเสพสื่อของคุณมีผลกระทบต่อการใช้ชีวิตประจำวันหรือไม่?'].map(impact_mapping)
df['ผลกระทบ_encoded'] = df['ผลกระทบ_encoded'].astype(float)
# ตรวจสอบผลลัพธ์
print(df[['คุณคิดว่าการเสพสื่อของคุณมีผลกระทบต่อการใช้ชีวิตประจำวันหรือไม่?', 'ผลกระทบ_encoded']].head())


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
# แปลงข้อมูลเป็น float64
df_selected['คะแนนรวม'] = df_selected['คะแนนรวม'].astype(float)
df['อายุ_encoded'] = df['อายุ_encoded'].astype(float)
df['ผลกระทบ_encoded'] = df['ผลกระทบ_encoded'].astype(float)

total_score = df_selected['คะแนนรวม'] 
age_encoded_custom = df['อายุ_encoded'] 
impact_encoded = df['ผลกระทบ_encoded'] 

# แยกข้อมูลเป็น features (X) และ target (y)
X = pd.DataFrame({
    'อายุ_encoded': age_encoded_custom,
    'คะแนนรวม': total_score
    })  # features
y = impact_encoded  # target



# แบ่งข้อมูลเป็น Training และ Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# สร้างโมเดล Random Forest
model = RandomForestClassifier(n_estimators=200, max_depth=12, max_features='sqrt', random_state=42)

# ฝึกโมเดล
model.fit(X_train, y_train)

# ทำนายผลลัพธ์
y_pred = model.predict(X_test)

# ประเมินผล
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# สร้าง confusion matrix
cm = confusion_matrix(y_test, y_pred)

# กำหนดลำดับใหม่ของ labels
labels_order = ["มีผลเสีย", "ไม่มีผลกระทบ", "ทั้งผลดีและผลเสีย", "มีผลดี"]

# สลับแถวและคอลัมน์ของ confusion matrix ให้ตรงกับลำดับ labels
index_order = [2, 0, 3, 1]  # ตำแหน่งของ labels ใหม่ในลำดับเดิม (0-based index)
cm_reordered = cm[np.ix_(index_order, index_order)]

# กลับลำดับของแกน Y (Actual) ให้เรียงจากล่างขึ้นบน
cm_reordered = cm_reordered[::-1]  # กลับด้านแถวของเมทริกซ์
labels_order_reversed = labels_order[::-1]  # กลับด้านของ labels บนแกน Y

# วาดกราฟ confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_reordered, annot=True, fmt="d", cmap="Blues", 
            xticklabels=labels_order, yticklabels=labels_order_reversed)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
